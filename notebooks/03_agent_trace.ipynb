# 03 — Agent Reasoning Trace Viewer

Visualise the ReAct Thought/Action/Observation loop for a single query.

```python
from chemvision.agent.trace import AgentStep, AgentTrace, StepType

# Manually construct a mock trace for UI prototyping
trace = AgentTrace(query='What functional groups are present?', image_path='sample.png')
trace.append(AgentStep(step_index=0, step_type=StepType.THOUGHT, content='I should use the molecular_structure skill.'))
trace.append(AgentStep(step_index=1, step_type=StepType.ACTION, content='molecular_structure', skill_name='molecular_structure'))
print(f'Steps: {trace.num_steps()}')
```
